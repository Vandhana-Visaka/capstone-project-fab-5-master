```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('output.csv')
```

```python
df = data[['ISBN','book_title','book_authors','book_desc','book_pages','genres','image_url']]
```

```python
df.head()
```

```python
data.head()
```

```python
df['language'] = 'English'
```

```python
df.head()
```

```python
day_list = pd.to_datetime(['2002-01-02','2010-05-05','2012-08-09','1988-01-01','1970-02-01'])
```

```python
country_list = ['USA','UK','USA']
```

```python
publisher_list = ['Penguin','Crossley','Harper Collins','Hachette','Macmillan','Random House','Simon and Schuster']
```

```python
df['country'] = np.random.choice(country_list, size=len(df))
```

```python
df['publisher'] = np.random.choice(publisher_list, size=len(df))
```

```python
df['publication_year'] = np.random.choice(day_list, size=len(df))
```

```python
df.head()
```

```python
df['book_pages'] = df['book_pages'].str.replace('pages', '')
```

```python
df['book_pages'] = df['book_pages'].fillna(0)
```

```python
df
```

```python
#ISBN,TITLE,SUMMARY,LANGUAGE,COUNTRY,NUM_PAGES,COVER,PUBLISHER,PUBLICATION_DATE,AUTHORS,GENRES
final = df[['ISBN','book_title','book_desc','language','country','book_pages','image_url','publisher','publication_year','book_authors','genres']]
```

```python
final
```

```python
final.to_csv(r'ultimate_books.csv',index=False)
```

```python
import pandas as pd
data = pd.read_csv('ultimate_books.csv',quotechar='"',skipinitialspace=True)
```

```python
tmp = list(data['ISBN'].unique())
indexes = []
for i in range(data.shape[0]):
    if data['ISBN'][i] in tmp:
        tmp.remove(data['ISBN'][i])
        indexes.append(i)
        
```

```python
new_data = data.iloc[indexes,:].reset_index(drop=True)
```

```python
new_data.columns
```

```python
new_data.iloc[:, 10].fillna('no_info', inplace=True)
```

```python
for i in range(new_data.shape[0]):
    if ',' in new_data.iloc[i, 10]:
        print(i)
```

```python
for i in range(new_data.shape[0]):
    new_data['book_desc'][i] = str(new_data['book_desc'][i])
    new_data['book_desc'][i] = new_data['book_desc'][i].replace(',', ';')
    new_data['book_title'][i] = new_data['book_title'][i].replace(',', ';')
```

```python
new_data.iloc[3890:3900,:].to_csv(r'ultimate_books_local.csv',index=False)
```
